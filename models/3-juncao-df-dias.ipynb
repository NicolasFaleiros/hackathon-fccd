```python
import pandas as pd
import gc
```

```python
# Concatenar dfs lances dos 3 dias

dia1 = pd.read_csv('arquivos_treino_teste\Formato_por_dia\df_lances_dia1.csv')
dia2 = pd.read_csv('arquivos_treino_teste\Formato_por_dia\df_lances_dia2.csv')
dia3 = pd.read_csv('arquivos_treino_teste\Formato_por_dia\df_lances_dia3.csv')
```

```python
dfs_lances_concatenados = pd.concat([dia1, dia2, dia3], axis=0)
```

```python
# Apagar as variáveis dia1, dia2 e dia3 da memória
del dia1, dia2, dia3

gc.collect()
```

```python
dfs_lances_concatenados.head(10)
```

```python
pd.set_option('display.max_rows', None)
dfs_lances_concatenados.isna().sum()
```

```python
df_treino = pd.read_csv('arquivos_base\\train.csv')
df_treino = df_treino.drop(columns=['conta_pagamento', 'endereco'])

df_treino_completo = df_treino.merge(dfs_lances_concatenados, on='id_participante', how='left')
df_treino_completo = df_treino_completo.dropna()
df_treino_completo.to_csv('arquivos_treino_teste\Formato_por_dia\df_treino_completo_3dias')
```

```python
df_teste = pd.read_csv('arquivos_base\\test.csv')
df_teste = df_teste.drop(columns=['conta_pagamento', 'endereco'])

df_teste_completo = df_teste.merge(dfs_lances_concatenados, on='id_participante', how='left')
df_teste_completo.to_csv('arquivos_treino_teste\Formato_por_dia\df_teste_completo_3dias.csv')
```

```python
df_teste_completo = df_teste_completo.drop_duplicates(subset='id_participante', keep='first')
df_teste_completo.shape
```
